```python
expenses = [
    {
        "id": 1,
        "date": "2026-05-01",
        "category": "식비",
        "amount": 8500,
        "memo": "김밥천국 점심"
    },
    {
        "id": 2,
        "date": "2026-05-01",
        "category": "교통",
        "amount": 1500,
        "memo": "버스"
    },
    {
        "id": 3,
        "date": "2026-05-02",
        "category": "카페",
        "amount": 4500,
        "memo": "아메리카노"
    }
]
```

```python
def add_expense(expenses):
    날짜=input('date를 입력해주세요 :')
    카테고리=input('category를 입력해주세요 :')
    총액=int(input('amount을 입력해 주세요 :'))
    메모= input('memo를 입력해주세요 :')
    if len(expenses) ==0:
        정보={'id': 1,'date':날짜,'category':카테고리,'amount':총액,'memo':메모}
    else:
        정보={'id':expenses[-1]['id']+1,'date':날짜,'category':카테고리,'amount':총액,'memo':메모}
    expenses.append(정보)


def show_expenses(expenses):
    if len(expenses) ==0:
        print('등록된 지출이 없습니다.')
    else:
        for i in range(len(expenses)):
            print(f"{expenses[i]['id']} | {expenses[i]['date']}|{expenses[i]['category']}|{expenses[i]['amount']}|{expenses[i]['memo']}")


def show_total_amount(expenses):
    total=0
    for i in range(len(expenses)):
        price=expenses[i]['amount']
        total+=price
    print(f'총액은 {total}원')

def show_category_summary(expenses):
    category_totals = {}

    for i in range(len(expenses)):
        category = expenses[i]["category"]
        amount = expenses[i]["amount"]

        if category in category_totals:
            category_totals[category] += amount
        else:
            category_totals[category] = amount

    for category, total in category_totals.items():
        print(f"{category}는 {total}원입니다.")


def search_by_date(expenses):
    날짜=input('날짜를 입력하세요 :')
    found=False
    for i in range(len(expenses)):
        if 날짜 == expenses[i]['date']:
            print(expenses[i])
            found=True
    
    if found ==False:
        print('해당 날짜의 지출이 없습니다.')

def delete_expense(expenses):
    show_expenses(expenses)
    delete_id=int(input('삭제할 id를 입력해주세요 :'))
    found=False
    for i in range(len(expenses)):
        if delete_id==expenses[i]['id']:
            found=True
            del expenses[i]
            print(f'{delete_id}가 삭제되었습니다.')
            break
    if found==False:
        print('삭제할 id가 없습니다.')



while True:
    choice = input('메뉴를 선택하세요:')

    if choice =='0':
        print('프로그램을 종료합니다')
        break
    elif choice =='1':
        print('지출 추가 기능입니다.')
        add_expense(expenses)
    elif choice =='2':
        print('지출 목록 보기입니다.')
        show_expenses(expenses)
    elif choice =='3':
        print('카테보기별 합계 보기입니다.')
        show_category_summary(expenses)
    elif choice =='4':
        print('날짜별 검색입니다.')
        search_by_date(expenses)
    elif choice =='5':
        print('지출 삭제 입니다.')
        delete_expense(expenses)
    elif choice =='6':
        print('전체 지출 합계 보기입니다.')
        show_total_amount(expenses)
    elif choice =='7':
        print('저장하기입니다.')
    else :
        print('잘못된 입력입니다.')
    
    




    
```

```python
def delete_expense(expenses):
    show_expenses(expenses)
    delete_id=int(input('삭제할 id를 입력해주세요 :'))
    found=False
    for i in range(len(expenses)):
        if delete_id==expenses[i]['id']:
            found=True
            del expenses[i]
            print(f'{delete_id}가 삭제되었습니다.')
            break
    if found==False:
        print('삭제할 id가 없습니다.')



```
