```python
import json
from datetime import datetime
# 순서가 중요함 
# 7.  저장하기
DATA_FILE = 'expenses.json'
def save_expenses(expenses):
    # 파일을 쓰기모드로 열기
    with open(DATA_FILE,'w',encoding='utf-8') as f: 
    # json.dump() # python 데이터를 json파일로 저장,리스트데이터파일을 저장/한글이 깨지지 않게 저장/ 보기좋게 들여쓰기
        json.dump(expenses,f,ensure_ascii=False,indent=4)
    print('저장되었습니다.')


def load_expenses():
    try:
        # f는 객체로 사용됨 
        with open(DATA_FILE,'r',encoding='utf-8') as f:
            return json.load(f)
    except FileNotFoundError:
        return []

expenses=load_expenses()


def show_menu():
    print()
    print("============ 지출 관리 프로그램 ============")
    print("1. 지출 추가")
    print("2. 지출 목록 보기")
    print("3. 카테고리별 합계 보기")
    print("4. 날짜별 검색")
    print("5. 지출 삭제")
    print("6. 전체 지출 합계 보기")
    print("7. 저장하기")
    print("0. 종료")
    print("==========================================")

#1. 지출 추가하기 
def add_expense(expenses):
    날짜=input('date를 입력해주세요 :')
    try:
        datetime.strptime(날짜,'%Y-%m-%d')
    except ValueError:
        print('날짜 형식은 YYYY-MM-DD 형식으로 입력하세요')
        return
    카테고리=input('category를 입력해주세요 :')
    try:
        총액=int(input('amount을 입력해 주세요 :'))
    except ValueError:
        print('숫자만 입력하세요.')
        return   # 함수를 바로 끝내버림  만약에 continue라면 for 문이나 while문 안에서 다시 처음의 while, for로 돌아감
    if 총액 <=0:
        print('금액은 1원 이상이어야 합니다.')
        return
    메모= input('memo를 입력해주세요 :')
    if len(expenses) ==0:
        정보={'id': 1,'date':날짜,'category':카테고리,'amount':총액,'memo':메모}
    else:
        id_list=[]
        for i in range(len(expenses)):
            id_list.append(expenses[i]['id'])
        id_l=max(id_list)+1
        정보={'id':id_l,'date':날짜,'category':카테고리,'amount':총액,'memo':메모}
    expenses.append(정보)
    print('지출이 추가되었습니다.')

#2. 전체 지출 보여주기 
def show_expenses(expenses):
    if len(expenses) ==0:
        print('등록된 지출이 없습니다.')
    else:
        for i in range(len(expenses)):
            print(f"{expenses[i]['id']} | {expenses[i]['date']}|{expenses[i]['category']}|{expenses[i]['amount']}|{expenses[i]['memo']}")

#3. 카테고리별 합계금액
def show_category_summary(expenses):
    category_totals = {}
    if len(expenses)==0:
        print('등록된 지출이 없습니다')
        return

    for i in range(len(expenses)):
        category = expenses[i]["category"]
        amount = expenses[i]["amount"]

        if category in category_totals:
            category_totals[category] += amount
        else:
            category_totals[category] = amount

    for category, total in category_totals.items():
        print(f"{category}는 {total}원입니다.")


#4. 날짜별 검색
def search_by_date(expenses):
    if len(expenses)==0:
        print('등록된 지출이 없습니다')
        return

    날짜=input('날짜를 입력하세요 :')
    try:
        datetime.strptime(날짜,'%Y-%m-%d')
    except ValueError:
        print('날짜 형식은 YYYY-MM-DD로 입력해야 합니다.')
        return
    found=False
    for i in range(len(expenses)):
        if 날짜 == expenses[i]['date']:
            print(f"{expenses[i]['id']} | {expenses[i]['date']}|{expenses[i]['category']}|{expenses[i]['amount']}|{expenses[i]['memo']}")

            found=True
    
    if found ==False:
        print('해당 날짜의 지출이 없습니다.')

#5. 삭제하기
def delete_expense(expenses):
    if len(expenses)==0:
        print('등록된 지출이 없습니다')
        return
    show_expenses(expenses)
    try:
        delete_id=int(input('삭제할 id를 입력해주세요 :'))
    except ValueError:
        print('id는 숫자로 입력해야합니다.')
        return
    found=False
    for i in range(len(expenses)):
        if delete_id==expenses[i]['id']:
            found=True
            del expenses[i]
            print(f'{delete_id}가 삭제되었습니다.')
            break
    if found==False:
        print('삭제할 id가 없습니다.')

#6. 전체 지출 합계보기
def show_total_amount(expenses):
    if len(expenses)==0:
        print('전체 지출 합계 : 0원')
        return
    total=0
    for i in range(len(expenses)):
        price=expenses[i]['amount']
        total+=price
    print(f'총액은 {total}원')



while True:
    show_menu()
    choice = input('메뉴를 선택하세요:')

    if choice =='0':
        save_expenses(expenses)
        print('프로그램을 종료합니다')
        break
    elif choice =='1':
        print('지출 추가 기능입니다.')
        add_expense(expenses)
    elif choice =='2':
        print('지출 목록 보기입니다.')
        show_expenses(expenses)
    elif choice =='3':
        print('카테보기별 합계 보기입니다.')
        show_category_summary(expenses)
    elif choice =='4':
        print('날짜별 검색입니다.')
        search_by_date(expenses)
    elif choice =='5':
        print('지출 삭제 입니다.')
        delete_expense(expenses)
    elif choice =='6':
        print('전체 지출 합계 보기입니다.')
        show_total_amount(expenses)
    elif choice =='7':
        print('저장하기입니다.')
        save_expenses(expenses)
    else :
        print('잘못된 입력입니다.')
    
```
